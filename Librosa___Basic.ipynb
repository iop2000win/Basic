```python
import os

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False

import librosa
```

```python
data_path = r'D:\Study\[Competition] English Sound Classification\Data\test'
src_path = os.listdir(data_path)[0]
```

```python
# y : 음원의 파형 데이터
# sr(samplig rate) : 주파수 분석 및 파형의 시간 간격을 결정
y, sr = librosa.load(os.path.join(data_path, src_path))
```

```python
print(y, len(y))
print(sr)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (16, 10))

ax.plot(y,
        color = 'g',
        alpha = 0.5)

ax.grid()
```

# STFT (Short Time Fourier Transform)

Time 도메인의 파형을 Frequency 도메인으로 변형시키는 푸레이 변환

```python
stft_result = librosa.stft(y, n_fft = 4096, win_length = 4096, hop_length = 512)
```
