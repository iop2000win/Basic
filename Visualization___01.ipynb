# 스타일

```python
import matplotlib
import matplotlib.pyplot as plt

# matplotlib option 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows, 폰트 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 사이즈 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 기호 깨짐 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, linewidth = 1,
         marker = '^',
         markersize = 15, # ms
         markerfacecolor = 'yellow', # mfc
         markeredgecolor = 'red', # mec
         linestyle = '-.', # :, --, -.,... ls
         alpha = 0.7, # 투명도 (0 ~ 1) >>> 중요한 정보 강조
         color = 'b') # 라인 색깔 설정
```

```python
import matplotlib
import matplotlib.pyplot as plt

# matplotlib option 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows, 폰트 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 사이즈 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 기호 깨짐 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b'] # 리스트 형태로 color 값 입력을 통해 bar 마다 다른 색깔 부여 가능

plt.bar(labels, values, color = colors, alpha = 0.5)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b'] # 리스트 형태로 color 값 입력을 통해 bar 마다 다른 색깔 부여 가능

plt.bar(labels, values,
        width = 0.5, # 막대의 두께를 조절
        color = colors,
        alpha = 0.5)

# 값의 전체적인 스케일이 비슷해서 차이를 눈으로 확인하기가 어려울 경우,
# 그래프의 표시 범위를 조정함으로써 시각화 효과를 극대화할 수 있다.
plt.ylim(175, 195)
plt.xticks(rotation = 45) # 축값 각도 조절 가능
plt.yticks(rotation = 45)

plt.show()
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
ticks = ['1번 학생', '2번 학생', '3번 학생']

plt.bar(labels, values)
plt.xticks(labels, ticks) # xticks 값의 표시 값만 바꿔서 표현 가능
```

# 막대 그래프 (심화)

```python
import matplotlib
import matplotlib.pyplot as plt

# matplotlib option 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows, 폰트 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 사이즈 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 기호 깨짐 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b'] # 리스트 형태로 color 값 입력을 통해 bar 마다 다른 색깔 부여 가능

plt.barh(labels, values, color = colors, alpha = 0.5)

plt.xlim(175, 200)
```

```python
bar = plt.bar(labels, values)

bar[0].set_hatch('//')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 200)

for idx, rect in enumerate(bar):
#     print(idx, rect)
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha = 'center', color = 'b')
```
