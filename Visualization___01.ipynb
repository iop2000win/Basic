# 0. figure and subplot

- figure: 서브플롯을 작성하는 영역 (Figure 클래스의 인스턴스로, 그래프를 작성하는 하나의 큰 틀이라고 생각하면 된다.)
- subplot: 그래프를 작성하기 위한 영역 (최소 한 개 이상의 서브플롯이 필요)

```python
fig = plt.figure()
ax = fig.add_subplot(111)

dat = [0, 1]
ax.plot(dat)

plt.show()
```

### add_subplot - subplot을 추가하는 방법

```python
fig = plt.figure() # Figure 클래스의 인스턴스 생성

ax1 = fig.add_subplot(2, 2, 1) # fig 인스턴스의 메서드 add_subplot을 이용해 subplot을 생성
ax2 = fig.add_subplot(2, 2, 2) # 2행 2열 2번을 의미. fig를 4등분하여, 좌상우하순으로 번호설정
ax3 = fig.add_subplot(2, 2, 3)

for i, ax in enumerate([ax1, ax2, ax3], start = 1):
    txt = f'ax{i}\n(22{i})'
    ax.text(0.2, 0.4, txt, fontsize = 24) # subplot에 텍스트를 출력하는 기능(x좌표, y좌표, 텍스트내용, 폰트사이즈)

plt.show()
```

### plt.subplots() - figure 생성 시에 subplots도 같이 생성하는 방식

```python
fig, axes = plt.subplots(2, 2)
print(type(axes))
print(axes)

for i in range(len(axes)):
    for j in range(len(axes)):
        txt = f'ax[{i}][{j}]'
        axes[i][j].text(0.2, 0.4, txt, fontsize = 24)
```

### GridSpec - subplots를 좀 더 유연하게 사용할 수 있도록 하는 기능

```python
from matplotlib.gridspec import GridSpec
```

```python
# 미리 그래프의 형태를 구상하고 세팅하는 경우
# 그래프의 배치가 균일한 간격 내에서 이루어지는 경우 
gs = GridSpec(3, 3)

fig = plt.figure(figsize = (5, 5))

ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0:2])
ax3 = fig.add_subplot(gs[1:, -1])
ax4 = fig.add_subplot(gs[2, 0])
ax5 = fig.add_subplot(gs[2, 1])

ax1.text(0.2, 0.4, f'ax[0, :]', fontsize = 12)
ax2.text(0.2, 0.5, f'ax[1, :-1]', fontsize = 12)
ax2.text(0.2, 0.3, f'ax[1, 0:2]', fontsize = 12)
ax3.text(0.2, 0.3, f'ax[1:, -1]', fontsize = 12)

fig.tight_layout()
```

```python
# https://zephyrus1111.tistory.com/226
# add_gridspec을 이용하여 좀 더 유연하게 서브플롯을 구성하는 방법
fig = plt.figure(figsize = (12, 9), facecolor = 'white')

gs1_nrows = 3
gs1_ncols = 2

gs1 = fig.add_gridspec(nrows = gs1_nrows,
                       ncols = gs1_ncols,
                       left = 0,
                       right = 0.47)

ax1_1 = fig.add_subplot(gs1[0, :])
ax1_2 = fig.add_subplot(gs1[1, :])
ax1_3 = fig.add_subplot(gs1[2, 0])
ax1_4 = fig.add_subplot(gs1[2, 1])

gs2_nrows = 2
gs2_ncols = 2

gs2 = fig.add_gridspec(nrows = gs2_nrows,
                       ncols = gs2_ncols,
                       left = 0.53,
                       right = 1)

ax2_1 = fig.add_subplot(gs2[0, 0])
ax2_2 = fig.add_subplot(gs2[0, 1])
ax2_3 = fig.add_subplot(gs2[1, 0])
ax2_4 = fig.add_subplot(gs2[1, 1])

plt.show()
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b'] # 리스트 형태로 color 값 입력을 통해 bar 마다 다른 색깔 부여 가능

plt.bar(labels, values,
        width = 0.5, # 막대의 두께를 조절
        color = colors,
        alpha = 0.5)

# 값의 전체적인 스케일이 비슷해서 차이를 눈으로 확인하기가 어려울 경우,
# 그래프의 표시 범위를 조정함으로써 시각화 효과를 극대화할 수 있다.
plt.ylim(175, 195)
plt.xticks(rotation = 45) # 축값 각도 조절 가능
plt.yticks(rotation = 45)

plt.show()
```

# 막대 그래프 (심화)

```python
import matplotlib
import matplotlib.pyplot as plt

# matplotlib option 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows, 폰트 설정
matplotlib.rcParams['font.size'] = 15 # 폰트 사이즈 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 기호 깨짐 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b'] # 리스트 형태로 color 값 입력을 통해 bar 마다 다른 색깔 부여 가능

plt.barh(labels, values, color = colors, alpha = 0.5)

plt.xlim(175, 200)
```

```python
bar = plt.bar(labels, values)

bar[0].set_hatch('//')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 200)

for idx, rect in enumerate(bar):
#     print(idx, rect)
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha = 'center', color = 'b')
```
